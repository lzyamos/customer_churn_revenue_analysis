```python
import pandas as pd
import numpy as np
import os

os.chdir("..")
np.random.seed(42)
num_customers = 5000

customer_ids = range(1, num_customers + 1)

customers = pd.DataFrame({
    "customer_id": customer_ids,
    "signup_date": pd.to_datetime("2021-01-01") + pd.to_timedelta(np.random.randint(0, 900, num_customers), unit="D"),
    "country": np.random.choice(["USA", "UK", "Canada", "Australia"], num_customers),
})

subscriptions = pd.DataFrame({
    "customer_id": customer_ids,
    "plan_type": np.random.choice(["Basic", "Standard", "Premium"], num_customers, p=[0.5, 0.3, 0.2]),
    "monthly_price": np.random.choice([20, 40, 70], num_customers),
})

usage = pd.DataFrame({
    "customer_id": customer_ids,
    "usage_hours": np.round(np.random.gamma(2, 5, num_customers), 2),
})

support = pd.DataFrame({
    "customer_id": customer_ids,
    "ticket_count": np.random.poisson(1.2, num_customers),
})

churn = pd.DataFrame({
    "customer_id": customer_ids,
    "churned": np.random.choice(["Yes", "No"], num_customers, p=[0.25, 0.75]),
})

payments = pd.DataFrame({
    "customer_id": customer_ids,
    "monthly_revenue": subscriptions["monthly_price"],
})

customers.to_csv("data/raw/customers.csv", index=False)
subscriptions.to_csv("data/raw/subscriptions.csv", index=False)
usage.to_csv("data/raw/usage.csv", index=False)
support.to_csv("data/raw/support_tickets.csv", index=False)
churn.to_csv("data/raw/churn.csv", index=False)
payments.to_csv("data/raw/payments.csv", index=False)
```
