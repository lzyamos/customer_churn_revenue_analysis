```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.chdir("..")
df = pd.read_csv("data/cleaned/customer_analysis_dataset.csv")
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
df["monthly_revenue"].describe()

# Revenue by plan
revenue_by_plan = df.groupby("plan_type")["monthly_revenue"].mean().reset_index()

sns.barplot(data=revenue_by_plan, x="plan_type", y="monthly_revenue")
plt.title("Average Monthly Revenue by Subscription Plan")
plt.xlabel("Plan Type")
plt.ylabel("Average Monthly Revenue")
plt.tight_layout()
plt.savefig("visuals/revenue_by_plan.png")
plt.show()

# Revenue tier destribution
sns.countplot(data=df, x="revenue_tier")
plt.title("Customer Distribution by Revenue Tier")
plt.xlabel("Revenue Tier")
plt.ylabel("Customer Count")
plt.tight_layout()
plt.savefig("visuals/revenue_tier_distribution.png")
plt.show()

# Usage vs revenue
sns.boxplot(data=df, x="usage_level", y="monthly_revenue")
plt.title("Monthly Revenue by Usage Level")
plt.xlabel("Usage Level")
plt.ylabel("Monthly Revenue")
plt.tight_layout()
plt.savefig("visuals/usage_vs_revenue.png")
plt.show()

# Revenue vs support load
sns.boxplot(data=df, x="support_intensity", y="monthly_revenue")
plt.title("Monthly Revenue by Support Intensity")
plt.xlabel("Support Intensity")
plt.ylabel("Monthly Revenue")
plt.tight_layout()
plt.savefig("visuals/revenue_vs_support.png")
plt.show()

# Churn context
sns.boxplot(data=df, x="churned", y="monthly_revenue")
plt.title("Monthly Revenue by Churn Status")
plt.xlabel("Churned")
plt.ylabel("Monthly Revenue")
plt.tight_layout()
plt.savefig("visuals/revenue_by_churn.png")
plt.show()

```
